## RETURN 데이터 가공

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import gmplot 
import statistics

from pandas import DataFrame
```

```python
train_rent = pd.read_csv('./data/rent/total_rent.csv')
train_rent.head(5)
```

```python
tashu = pd.read_csv('./data/clean/tashu.csv')
tashu
```

```python
rt_clean = train_rent.drop(columns=['WEEKDAY', 'SEASON', 'TEMPERATURE','RAINFALL',
                                      'WINDSPEED','HUMLDITY','SNOWFALL','RENTCOUNT'])
rt = tashu.rename(columns = {'RETURN_STATION': 'STATION', 'RETURN_YEAR': 'YEAR', 
                                        'RETURN_MONTH': 'MONTH', 'RETURN_DAY': 'DAY', 'RETURN_HOUR': 'HOUR'})
```

```python
rt
```

```python
rt_join = pd.merge(rt_clean, rt, how='outer')
```

```python
rt_join = rt_join.dropna()
```

```python
rt_join = rt_join[rt_join.STATION < 145]
```

```python
rt_join = rt_join[rt_join.RENT_STATION < 145]
```

```python
rt_join = rt_join[rt_join.YEAR < 2016]
```

```python
rt_join
```

```python
temp = rt_join.groupby(['STATION','YEAR','MONTH','DAY','HOUR']).size().to_frame('RETURNCOUNT').reset_index()
temp
```

```python
train_rent = train_rent.drop(columns=['RENTCOUNT'])
```

```python
train_rent = pd.merge(train_rent, temp, how='outer')
```

```python
train_rent = train_rent.replace({'RETURNCOUNT': np.nan}, {'RETURNCOUNT': 0.0})
```

```python
train_rent
```

```python
train_rent.to_csv('total_return.csv',index=False)
```
