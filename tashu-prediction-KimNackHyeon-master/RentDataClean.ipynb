## RENT 데이터 가공

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import gmplot 
import statistics

from pandas import DataFrame
```

```python
train_rent = pd.read_csv('./data/rent/total_rent.csv')
train_rent.head(7)
# len(train_rent)
```

```python
tashu = pd.read_csv('./data/clean/tashu.csv')
# len(rent)
tashu.head(7)
```

```python
rent_clean = train_rent.drop(columns=['WEEKDAY', 'SEASON', 'TEMPERATURE','RAINFALL',
                                      'WINDSPEED','HUMLDITY','SNOWFALL'])

rent = tashu.rename(columns = {'RENT_STATION': 'STATION', 'RENT_YEAR': 'YEAR', 
                                        'RENT_MONTH': 'MONTH', 'RENT_DAY': 'DAY', 'RENT_HOUR': 'HOUR'})
```

```python
rent_join = pd.merge(rent_clean, rent, how='outer')
```

```python
rent_join = rent_join.dropna()
```

```python
rent_join
```

```python
temp = rent_join.groupby(['STATION','YEAR','MONTH','DAY','HOUR']).size().to_frame('RENTCOUNT').reset_index()
```

```python
train_rent = train_rent.drop(columns=['RENTCOUNT'])
```

```python
train_rent = pd.merge(train_rent, temp, how='outer')
```

```python
train_rent = train_rent.replace({'RENTCOUNT': np.nan}, {'RENTCOUNT': 0.0})
```

```python
train_rent
```

```python
train_rent.to_csv('total_rent.csv',index=False)
```
